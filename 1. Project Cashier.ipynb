```python
List_Menu =""

List = '''' Selamat Datang di Pasar Buah 
        List Menu : 
      1. Menampilkan Daftar Buah 
      2. Menambah Buah 
      3. Menghapus Buah 
      4. Membeli buah 
      5. Exit Program  '''

while List_Menu != 5: 
    Daftar = [{ 'Nama': "Apel", "Stock": 20,"Harga":10000},
            { 'Nama': "Jeruk", "Stock": 15,"Harga":15000},
            { 'Nama': "Anggur", "Stock": 25,"Harga":20000},]

    print (List)

    List_Menu = int(input("Masukkan Perintah"))

    if List_Menu == 1:
        print ('\n')
        print ('\n')
        print ('Index' '\t' 'Nama' '\t' 'Stock' '\t' "Harga")

        for i in range ( 1, len(Daftar)+1) :
            print ( f'{i-1}{'\t'}{Daftar[i-1]["Nama"]}{'\t'}{Daftar[i-1]["Stock"]}{'\t'}{Daftar[i-1]["Harga"]}')
    elif List_Menu == 2:
        print ('\n')
        print ('\n')
        

        Buah_tambahan = (input("Masukkan nama buah"))
        Stock_tambahan = int(input("Masukkan stock buah"))
        Harga_Buah = int(input("Masukkan harga buah"))


        Daftar.append ({'Nama' : Buah_tambahan, 'Stock' : Stock_tambahan, 'Harga' : Harga_Buah})

        print ('Index' '\t' 'Nama' '\t' 'Stock' '\t' "Harga")
        for i in range ( 1, len(Daftar)+1) :
            print ( f'{i-1}{'\t'}{Daftar[i-1]["Nama"]}{'\t'}{Daftar[i-1]["Stock"]}{'\t'}{Daftar[i-1]["Harga"]}')

    elif List_Menu == 3:
        print ('\n')
        print ('\n')

        Hapus_Buah = int(input('Masukkan index buah'))


        Daftar.pop(Hapus_Buah) 
        print ('Index' '\t' 'Nama' '\t' 'Stock' '\t' "Harga")
        for i in range ( 1, len(Daftar)+1) :
            print ( f'{i-1}{'\t'}{Daftar[i-1]["Nama"]}{'\t'}{Daftar[i-1]["Stock"]}{'\t'}{Daftar[i-1]["Harga"]}')

    elif List_Menu == 4 :
        print ('\n')
        chart = []
        print ('\n')
        print ('\n')
        print ('Index' '\t' 'Nama' '\t' 'Stock' '\t' "Harga")
        for i in range ( 1, len(Daftar)+1) :
            print ( f'{i-1}{'\t'}{Daftar[i-1]["Nama"]}{'\t'}{Daftar[i-1]["Stock"]}{'\t'}{Daftar[i-1]["Harga"]}')
            beli_lagi = True 
        while beli_lagi:
                
            Index_Buah = int(input("Masukkan index buah"))
            if Index_Buah not in range(0, len(Daftar)):
                print("Index tidak valid") 
            Jumlah_Buah = int(input("Masukkan total buah"))
            Nama_buah = Daftar[Index_Buah]['Nama']
            Stock_buah = Daftar[Index_Buah]['Stock']

            if Jumlah_Buah > Stock_buah :
                print(f'Stock tidak mencukupi ')

            else : 
                Total_harga = Daftar[Index_Buah]['Harga'] * Jumlah_Buah
                data_beli = {"Nama": Nama_buah, "Qty" : Jumlah_Buah, "Harga" : Total_harga}
                chart.append(data_beli)
                print ('Nama' '\t' 'QTY' '\t' "Harga")
                for i in range (1, len(chart)+1) :
                    print ( f"{chart[i-1]['Nama']}\t{chart[i-1]['Qty']}\t{chart[i-1]['Harga']}")
                beli_lain = input("beli yang lain tidak ya/tidak")
                
                if beli_lain == "tidak" :
                    print("daftar belanja")
                    print ('Nama' '\t' 'QTY' '\t' "Harga")
                    result = 0 
                    for i in range (1, len(chart)+1) :
                        result = result + chart[i-1]['Harga']
                    for i in range (1, len(chart)+1) :
                        print ( f"{chart[i-1]['Nama']}\t{chart[i-1]['Qty']}\t{chart[i-1]['Harga']}")
                    print (f'harga total : {result}')
                    
                    pembayaran = True
                    while pembayaran:
                        print ('\n')
                        print ('\n')
                        print ('Masukkan uang anda')    
                        Masukkan_pembayaran = int(input("masukkan uang"))
                        if Masukkan_pembayaran < 0 :
                            print ("uang tidak boleh kurang dari 0")
                            
                        elif Masukkan_pembayaran == 0 :
                            print ("uang tidak boleh 0")
                            
                        elif Masukkan_pembayaran < result :
                            print ("uang kurang")
                            
                        elif Masukkan_pembayaran == result :
                            print ("terimakasih")
                            List_Menu = '5'
                            pembayaran = False
                            beli_lagi = False
                            
                        elif Masukkan_pembayaran > result :
                            Kembalian = Masukkan_pembayaran - result
                            print (f'''kembalian anda {Kembalian}
                            terimakasih''')
                            List_Menu = '5'
                            pembayaran = False
                            beli_lagi = False

















```
